```python
import pandas as pd
data = pd.read_csv('tgdd_product_description_grouped_preprocessed.csv')
```

```python
from keybert import KeyBERT
kw_model = KeyBERT(model="paraphrase-multilingual-MiniLM-L12-v2")
```

```python
label_words = []
for dataDes in data['description']:
  try:
    keywords = 250
    a_k = kw_model.extract_keywords(dataDes, top_n=keywords, use_mmr=True)
    a = []
    for word in a_k:
      a.append(word[0])
    
    label_words.append(a)
  except Exception as e: 
    print(e)
    break
```

```python
data['label']=label_words
```

```python
#Dữ liệu huấn luyện đầu vào
x_corpus = [ i.split() for i in data['description']]
```

```python
#Output dữ liện huấn luyện
y_label = []
for i in range(len(data['description'])):
  label = [1 if j in label_words[i] else 0 for j in data['description'][i].split()]
  y_label.append(label)
```

```python
import random
import numpy as np
t_1 = random.choice(x_corpus)
print(np.array(t_1))
```

```python
data['x_corpus'] = x_corpus
data['y_label'] = y_label
```

```python
different_length = [1 if len(input) != len(output) else 0 for input, output in zip(x_corpus, y_label)]
print("{} câu có độ dài in-out khác nhau.".format(sum(different_length)))
```
